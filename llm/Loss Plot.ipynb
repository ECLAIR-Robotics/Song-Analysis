# Loss Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
data = pd.read_csv('loss.csv')
```

```python
data[['train_loss', 'val_loss']].plot(xlabel='Epoch', ylabel='Loss')
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['epoch'], data['train_loss'])
plt.plot(data['epoch'], data['val_loss'])
plt.legend(['Train Loss', 'Validation Loss'])
plt.title('Model Loss over Fine-Tuning')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
